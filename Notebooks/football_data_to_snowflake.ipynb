```python
def wins(r):
    if r['winner'] == 'DRAW':
        return 'D'
    if r['winner'] == 'HOME_TEAM':
        if r['home_team_code'] == 'BVB':
            return 'W'
        return 'L'
    if r['winner'] == 'AWAY_TEAM':
        if r['away_team_code'] == 'BVB':
            return 'W'
        return 'L'
```

```python
code_series = team_hist_df.groupby('team_name').agg(pd.Series.mode)['away_team_code']
code_series
```

```python
team_hist_df = team_hist_df.set_index('team_name').merge(code_series, left_index=True, right_index=True).reset_index().rename(columns={'away_team_code_y':'team_code'})
team_hist_df
```

```python
team_hist_df = team_hist_df[team_hist_df['team_name'] != 'Galatasaray']
```

```python
def wins(r):
    if r['winner'] == 'DRAW':
        return 'D'
    if r['winner'] == 'HOME_TEAM':
        if r['home_team_code'] == r['team_code']:
            return 'W'
        return 'L'
    if r['winner'] == 'AWAY_TEAM':
        if r['away_team_code_x'] == r['team_code']:
            return 'W'
        return 'L'
```

```python
team_hist_df['wins']=team_hist_df.apply(wins, axis=1)
team_hist_df
```

```python
team_hist_df.drop(columns=['winner', 'away_team', 'away_team_code_x', 'home_team', 'home_team_code'], inplace=True)
team_hist_df
```
