```python
# API call
import requests

# Endpoint
url = "https://pro-api.coinmarketcap.com/v1/cryptocurrency/category"

# Headers for the request
headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': 'de7da518-a4aa-41d6-8c2c-d51fc2e14b6a',
}

# Query parameters
params = {
    'id': '6363a6c9cd197958bb543bf0'
}

# Make the API request
response = requests.get(url, headers=headers, params=params)

# Check the response
if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Error {response.status_code}: {response.text}")
```

```python
import pandas as pd
# Convert the 'data' key from the JSON response into a DataFrame
df = pd.DataFrame(data['data'])
# Expand the dictionary inside the coins column 
coins_df = pd.DataFrame(df['coins'].tolist())

#function to flatten json files contained in cells.
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        else:
            out[name[:-1]] = x

    flatten(y)
    return out

# Apply the flatten_json function to the 'quote' column
flattened_quotes = coins_df['quote'].apply(flatten_json)

# Convert the Series of dictionaries into a DataFrame
flattened_quote_df = pd.DataFrame(flattened_quotes.tolist())

# Concatenate the original coins_df and the flattened_quote_df
result_df = pd.concat([coins_df.drop(columns=['quote']), flattened_quote_df], axis=1)
```
